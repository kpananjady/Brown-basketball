```python
def automate_scraping():
    # function that feeds selenium different button options
    list_info=[]
    for drop in list_dropdown:
        #select the dropdown option
        driver.find_element_by_css_selector(drop).click()
        #click go 
        print(driver.find_element_by_css_selector(drop).text)
        roster = driver.find_element_by_css_selector(drop).text
        try:
            driver.find_element_by_css_selector('#main-content > article > div:nth-child(4) > header > div.sidearm-roster-select.sidearm-common-header-select.flex.large-6.columns > ul > li.sidearm-roster-select-year.flex.flex-item-1 > button').click()
        except:
            #on older rosters the CSS selector for the drop down menu changes 
            driver.find_element_by_css_selector('#main-content > article > div:nth-child(3) > header > div.sidearm-roster-select.sidearm-common-header-select.flex.large-6.columns > ul > li.sidearm-roster-select-year.flex.flex-item-1 > button').click()

        list_li = driver.find_elements_by_tag_name("li")
        for item in list_li:
            if item != '':
                list_info += [item.text + roster]
    return list_info        
```
